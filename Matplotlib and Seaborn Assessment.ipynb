# Plotting Matplotlib and seaborn

```python
# Question 1. Plotting graph For IRIS Dataset Using Seaborn Library And matplotlib.pyplot library
# A. Creating and customizing line charts using Matplotlib
# B. Visualizing relationships between two or more variables using scatter plots
# C. Studying distributions of variables using histograms & bar charts to
# D. Visualizing two-dimensional data using heatmaps
# E. Displaying images using Matplotlib's plt.imshow
# F. Plotting multiple Matplotlib and Seaborn charts in a grid

import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

iris = sns.load_dataset('iris')
iris
```

```python
# A. Creating and customizing line charts using Matplotlib

plt.plot(iris.index, iris["sepal_length"], "p--") 
plt.show 
```

```python
# B. Visualizing relationships between two or more variables using scatter plots

plt.title('Sepal Dimensions')
sns.scatterplot(x='sepal_length', 
                y='sepal_width', 
                hue='species',
                s=50,
                data=iris);
```

```python
# C. Studying distributions of variables using histograms & bar charts to

setosa_df = iris[iris.species == 'setosa']
versicolor_df = iris[iris.species == 'versicolor']
virginica_df = iris[iris.species == 'virginica']
plt.title('Distribution of Sepal Width')
plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], 
         bins=np.arange(2, 5, 0.25), alpha=0.4,
         stacked=True);             ## stacked True is used in this situation

plt.legend(['Setosa', 'Versicolor', 'Virginica']);

```

```python
# D. Visualizing two-dimensional data using heatmaps

plt.figure(figsize=(8, 6))
sns.heatmap(iris.corr(), annot=True, cmap='Reds')
```

```python
# E. Displaying images using Matplotlib's plt.imshow

from PIL import Image
img = Image.open('graph image.png')
plt.imshow(img);
```

```python
# F. Plotting multiple Matplotlib and Seaborn charts in a grid

fig, axs = plt.subplots(2, 2, figsize=(12, 8))
sns.boxplot(x='species', y='sepal_length', data=iris, ax=axs[0, 0])
sns.boxplot(x='species', y='sepal_width', data=iris, ax=axs[0, 1])
sns.boxplot(x='species', y='petal_length', data=iris, ax=axs[1, 0])
sns.boxplot(x='species', y='petal_width', data=iris, ax=axs[1, 1])

plt.tight_layout()
plt.show()
```
